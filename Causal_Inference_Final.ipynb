# **Simple ATE Calculation (Rubin-Neyman Potential Outcomes)**

```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from scipy.spatial.distance import cdist

# Load the data
df = pd.read_csv('df_updated.csv')

# 1. Simple ATE Calculation (Rubin-Neyman Potential Outcomes)

def calculate_simple_ate(data):
    treated = data[data['year'] >= 2023]['sale_value_nis']
    control = data[data['year'] < 2023]['sale_value_nis']

    ate = treated.mean() - control.mean()

    # Calculate standard error and confidence interval
    se = np.sqrt(treated.var() / len(treated) + control.var() / len(control))
    ci = stats.t.interval(0.95, len(treated) + len(control) - 2, loc=ate, scale=se)

    return ate, ci

# Calculate simple ATE for each distance group
distance_groups = {'0-500m': '0-500', '500-1000m': '500-1000', '1000m+': '1000+'}

print("Simple ATE Calculations (Rubin-Neyman Potential Outcomes):")
for group_name, group_code in distance_groups.items():
    group_data = df[df['distance_group'] == group_code]
    ate, ci = calculate_simple_ate(group_data)
    print(f"{group_name}: ATE = {ate:.2f}, 95% CI = ({ci[0]:.2f}, {ci[1]:.2f})")

print("\n" + "="*50 + "\n")
```

# **Pre-Metro Group Comparison Analysis (2021-2022)**

```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the data
df = pd.read_csv('df_updated.csv')

# Filter for pre-metro period (2021-2022)
df_pre_metro = df[df['year'].isin([2021, 2022])]

def compare_groups(data, distance_groups, numeric_features, categorical_features):
    results = {}
    for feature in numeric_features:
        feature_data = []
        for group_name, group_code in distance_groups.items():
            group_values = data[data['distance_group'] == group_code][feature]
            feature_data.append({
                'group': group_name,
                'mean': group_values.mean(),
                'median': group_values.median(),
                'std': group_values.std()
            })
        results[feature] = pd.DataFrame(feature_data)

    for feature in categorical_features:
        feature_data = []
        for group_name, group_code in distance_groups.items():
            value_counts = data[data['distance_group'] == group_code][feature].value_counts(normalize=True)
            feature_data.append({
                'group': group_name,
                'most_common': value_counts.index[0],
                'most_common_pct': value_counts.iloc[0] * 100
            })
        results[feature] = pd.DataFrame(feature_data)

    return results

# Define distance groups
distance_groups = {'0-500m': '0-500', '500-1000m': '500-1000', '1000m+': '1000+'}

# Define features for comparison
numeric_features = ['year_of_construction', 'bedrooms', 'sale_value_nis']
categorical_features = ['nature', 'city', 'closest_direct_address']

# Perform group comparison
comparison_results = compare_groups(df_pre_metro, distance_groups, numeric_features, categorical_features)

# Print comparison results
print("Pre-Metro Group Comparability Analysis (2021-2022):")
print("="*50)

for feature, result_df in comparison_results.items():
    print(f"\nFeature: {feature}")
    print(result_df)
    print("-"*50)

# Statistical tests for numeric features
print("\nStatistical Tests for Numeric Features:")
print("="*50)

for feature in numeric_features:
    print(f"\nFeature: {feature}")
    groups = [df_pre_metro[df_pre_metro['distance_group'] == code][feature] for code in distance_groups.values()]
    f_statistic, p_value = stats.f_oneway(*groups)
    print(f"One-way ANOVA: F-statistic = {f_statistic:.4f}, p-value = {p_value:.4f}")

# Chi-square tests for categorical features
print("\nChi-square Tests for Categorical Features:")
print("="*50)

for feature in categorical_features:
    print(f"\nFeature: {feature}")
    contingency_table = pd.crosstab(df_pre_metro['distance_group'], df_pre_metro[feature])
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
    print(f"Chi-square test: chi2 = {chi2:.4f}, p-value = {p_value:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.neighbors import NearestNeighbors



# Load the data
df = pd.read_csv('df_updated.csv')

def perform_psm(data, treatment_col, features):
    # Prepare the data for matching
    X = data[features]
    y = data[treatment_col]

    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Estimate propensity scores
    logistic = LogisticRegression(random_state=42)
    logistic.fit(X_scaled, y)
    propensity_scores = logistic.predict_proba(X_scaled)[:, 1]

    # Perform matching
    nn = NearestNeighbors(n_neighbors=1, metric='euclidean')
    nn.fit(propensity_scores[y == 0].reshape(-1, 1))

    distances, indices = nn.kneighbors(propensity_scores[y == 1].reshape(-1, 1))

    matched_control_indices = data[y == 0].index[indices.flatten()]
    matched_treated_indices = data[y == 1].index

    matched_data = pd.concat([
        data.loc[matched_treated_indices],
        data.loc[matched_control_indices]
    ])

    return matched_data

def assess_balance(data, features, treatment_col):
    balance = {}
    for feature in features:
        treated = data[data[treatment_col] == 1][feature]
        control = data[data[treatment_col] == 0][feature]
        balance[feature] = {
            'std_diff': (treated.mean() - control.mean()) / np.sqrt((treated.var() + control.var()) / 2),
            'var_ratio': treated.var() / control.var() if control.var() != 0 else np.inf
        }
    return pd.DataFrame(balance).T

def analyze_group(data, distance_group):
    # Define treatment (after metro opening)
    data['treatment'] = (data['year'] >= 2023).astype(int)

    # Select relevant features for matching
    features = ['year_of_construction', 'bedrooms']

    # Assess balance before matching
    balance_before = assess_balance(data, features, 'treatment')

    # Perform PSM
    matched_data = perform_psm(data, 'treatment', features)

    # Assess balance after matching
    balance_after = assess_balance(matched_data, features, 'treatment')

    print(f"\nBalance for {distance_group} group before matching:")
    print(balance_before)
    print(f"\nBalance for {distance_group} group after matching:")
    print(balance_after)

    # Calculate average treatment effect
    ate = matched_data[matched_data['treatment'] == 1]['sale_value_nis'].mean() - \
          matched_data[matched_data['treatment'] == 0]['sale_value_nis'].mean()

    print(f"\nAverage Treatment Effect for {distance_group}: {ate:.2f}")

    return ate, matched_data

# Perform analysis for each distance group
distance_groups = {
    '0-500m': '0-500',
    '500-1000m': '500-1000',
    '1000m+': '1000+'
}

results = {}
for group_name, group_code in distance_groups.items():
    group_data = df[df['distance_group'] == group_code]
    results[group_name] = analyze_group(group_data, group_name)

# Compare effects across groups
print("\nComparison of Average Treatment Effects:")
for group_name, (ate, _) in results.items():
    print(f"{group_name}: {ate:.2f}")
```

# **Mahalanobis Distance Matching**




```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from scipy.spatial.distance import cdist
# Load the data
df = pd.read_csv('df_updated.csv')

def perform_mahalanobis_matching(data, treatment_col, numeric_features):
    # Prepare the data for matching
    X = data[numeric_features]
    y = data[treatment_col]

    # Impute missing values and scale the features
    imputer = SimpleImputer(strategy='median')
    scaler = StandardScaler()

    X_processed = scaler.fit_transform(imputer.fit_transform(X))

    # Calculate covariance matrix
    cov = np.cov(X_processed.T)

    # Ensure covariance matrix is positive definite
    cov = np.linalg.pinv(cov)

    # Calculate Mahalanobis distances
    treated = X_processed[y == 1]
    control = X_processed[y == 0]

    distances = cdist(treated, control, metric='mahalanobis', VI=cov)

    # Find closest match for each treated unit
    matches = distances.argmin(axis=1)

    # Create matched dataset
    treated_indices = data[y == 1].index
    control_indices = data[y == 0].index[matches]

    matched_data = pd.concat([
        data.loc[treated_indices],
        data.loc[control_indices]
    ])

    return matched_data

def assess_balance(data, numeric_features, treatment_col):
    balance = {}

    for feature in numeric_features:
        treated = data[data[treatment_col] == 1][feature]
        control = data[data[treatment_col] == 0][feature]
        balance[feature] = {
            'std_diff': (treated.mean() - control.mean()) / np.sqrt((treated.var() + control.var()) / 2),
            'var_ratio': treated.var() / control.var() if control.var() != 0 else np.inf
        }

    return pd.DataFrame(balance).T

def analyze_group(data, distance_group):
    # Create a copy of the data to avoid SettingWithCopyWarning
    data = data.copy()

    # Define treatment (after metro opening)
    data['treatment'] = (data['year'] >= 2023).astype(int)

    # Select relevant features for matching
    numeric_features = ['year_of_construction', 'bedrooms']

    # Ensure numeric features are actually numeric
    for feature in numeric_features:
        data[feature] = pd.to_numeric(data[feature], errors='coerce')

    # Drop rows with NaN values after conversion
    data = data.dropna(subset=numeric_features)

    # Assess balance before matching
    balance_before = assess_balance(data, numeric_features, 'treatment')

    # Perform Mahalanobis matching
    matched_data = perform_mahalanobis_matching(data, 'treatment', numeric_features)

    # Assess balance after matching
    balance_after = assess_balance(matched_data, numeric_features, 'treatment')

    print(f"\nBalance for {distance_group} group before matching:")
    print(balance_before)
    print(f"\nBalance for {distance_group} group after matching:")
    print(balance_after)

    # Calculate average treatment effect
    ate = matched_data[matched_data['treatment'] == 1]['sale_value_nis'].mean() - \
          matched_data[matched_data['treatment'] == 0]['sale_value_nis'].mean()

    print(f"\nAverage Treatment Effect for {distance_group}: {ate:.2f}")

    return ate, matched_data

# Perform analysis for each distance group
distance_groups = {
    '0-500m': '0-500',
    '500-1000m': '500-1000',
    '1000m+': '1000+'
}

results = {}
for group_name, group_code in distance_groups.items():
    group_data = df[df['distance_group'] == group_code]
    results[group_name] = analyze_group(group_data, group_name)

# Compare effects across groups
print("\nComparison of Average Treatment Effects:")
for group_name, (ate, _) in results.items():
    print(f"{group_name}: {ate:.2f}")
```

## **Mahalanobis Distance Matching with prices normalization**




```python
import pandas as pd
import numpy as np
from scipy import stats
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from scipy.spatial.distance import cdist

# Load the data
df = pd.read_csv('df_updated.csv')

# Create price index adjustment factor
# Using 2021 as the base year (index = 1.00)
price_index = {
    2021: 1.00,
    2022: 1.147,  # 14.7% increase from 2021 to 2022
    2023: 1.147 * 0.99,  # 1% decrease from 2022 to 2023
    2024: 1.147 * 0.99 * 1.053  # 5.3% increase from 2023 to 2024
}

# Adjust sale prices based on the year
df['adjusted_sale_value_nis'] = df.apply(lambda row: row['sale_value_nis'] / price_index[row['year']], axis=1)

# Print unique years in the dataset for verification
print("Unique years in the dataset:", df['year'].unique())

def perform_mahalanobis_matching(data, treatment_col, numeric_features, categorical_features):
    X = data[numeric_features + categorical_features]
    y = data[treatment_col]

    # Preprocessing
    numeric_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])

    categorical_transformer = Pipeline(steps=[
        ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    preprocessor = ColumnTransformer(
        transformers=[
            ('num', numeric_transformer, numeric_features),
            ('cat', categorical_transformer, categorical_features)
        ])

    # Fit and transform the data
    X_processed = preprocessor.fit_transform(X)

    # Ensure X_processed is a dense array
    X_dense = X_processed.toarray() if hasattr(X_processed, 'toarray') else X_processed

    # Calculate Mahalanobis distances
    cov = np.cov(X_dense.T)
    cov_inv = np.linalg.pinv(cov)

    treated = X_dense[y == 1]
    control = X_dense[y == 0]

    distances = cdist(treated, control, metric='mahalanobis', VI=cov_inv)

    matches = distances.argmin(axis=1)

    matched_data = pd.concat([
        data[y == 1],
        data[y == 0].iloc[matches]
    ])

    return matched_data

def calculate_ate(matched_data, treatment_col='treatment'):
    treated = matched_data[matched_data[treatment_col] == 1]['adjusted_sale_value_nis']
    control = matched_data[matched_data[treatment_col] == 0]['adjusted_sale_value_nis']

    ate = treated.mean() - control.mean()
    se = np.sqrt(treated.var() / len(treated) + control.var() / len(control))
    ci = stats.t.interval(0.95, len(treated) + len(control) - 2, loc=ate, scale=se)

    return ate, ci

# Define distance groups
distance_groups = {'0-500m': '0-500', '500-1000m': '500-1000', '1000m+': '1000+'}

# Define features for matching
numeric_features = ['year_of_construction', 'bedrooms']
categorical_features = ['nature', 'city', 'closest_direct_address']

results = {}

print("Mahalanobis Matching Results (with price adjustment, 2021 base year):")
for group_name, group_code in distance_groups.items():
    group_data = df[df['distance_group'] == group_code].copy()
    group_data['treatment'] = (group_data['year'] >= 2023).astype(int)

    matched_data = perform_mahalanobis_matching(group_data, 'treatment', numeric_features, categorical_features)

    ate, ci = calculate_ate(matched_data)
    results[group_name] = (ate, ci)
    print(f"{group_name}: ATE = {ate:.2f}, 95% CI = ({ci[0]:.2f}, {ci[1]:.2f})")

print("\n" + "="*50 + "\n")

# Comparison of Results

print("Comparison of Mahalanobis Matching Results (with price adjustment, 2021 base year):")
print("Distance   | ATE           | 95% CI")
print("-"*40)

for group, (ate, ci) in results.items():
    print(f"{group:<11}| {ate:>11.2f} | ({ci[0]:>6.2f}, {ci[1]:>6.2f})")

# Add discussion and interpretation of results here
```

# Regression Discontinuity Design (RDD)

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.utils import resample

# Load the data
df = pd.read_csv('df_updated.csv')

# Create price index adjustment factor
# Using 2021 as the base year (index = 1.00)
price_index = {
    2021: 1.00,
    2022: 1.147,  # 14.7% increase from 2021 to 2022
    2023: 1.147 * 0.99,  # 1% decrease from 2022 to 2023
    2024: 1.147 * 0.99 * 1.053  # 5.3% increase from 2023 to 2024
}

# Filter data based on area and percentage of sale
df = df[df['area'] > 2]
df = df[df['percentage_of_sale'] > 0.2]

# Adjust sale prices based on the year
df['adjusted_sale_value_nis'] = df.apply(lambda row: (row['sale_value_nis']) / (row['percentage_of_sale'] * price_index[row['year']] * row['area']), axis=1)

# Filter the data for the specified years (2021-2023)
df_pre_metro = df[(df['year'].isin([2021, 2022, 2023])) & (df['sale_date'] < '2023-08-17')].copy()

# Define distance groups
df_pre_metro.loc[:, 'distance_group'] = pd.cut(
    df_pre_metro['direct_distance_km'],
    bins=[0, 0.5, 1, float('inf')],
    labels=['0-0.5 km', '0.5-1 km', 'above 1 km']
)

# Assign treatment codes based on distance group
df_pre_metro.loc[:, 'treatment'] = df_pre_metro['distance_group'].map({
    '0-0.5 km': 0,
    '0.5-1 km': 1,
    'above 1 km': 2
})

# Regression Discontinuity Analysis with Covariate Adjustment and Bootstrap ATE
def regression_discontinuity_analysis(data, distance_col, price_col, numeric_features, n_bootstraps=1000):
    # Compute descriptive statistics for each group
    group_stats = data.groupby('distance_group')[price_col].agg(['mean', 'median', 'std']).reset_index()
    print("Group Statistics:")
    print(group_stats)

    # Perform ANOVA for differences among the three groups
    prices_0_05km = data[data['treatment'] == 0][price_col]
    prices_05_1km = data[data['treatment'] == 1][price_col]
    prices_above_1km = data[data['treatment'] == 2][price_col]
    f_stat, p_value = stats.f_oneway(prices_0_05km, prices_05_1km, prices_above_1km)
    print("\nANOVA test for price difference across distance groups:")
    print(f"F-statistic = {f_stat:.4f}, p-value = {p_value:.4f}")

    # Prepare data for regression with dummies for treatment groups
    X = pd.get_dummies(data[['treatment'] + numeric_features], drop_first=True)
    y = data[price_col]

    # Ensure all columns in X and y are numeric and drop NaN values
    X = X.apply(pd.to_numeric, errors='coerce').dropna()
    y = pd.to_numeric(y, errors='coerce').loc[X.index]

    # Add a constant for the intercept
    X = sm.add_constant(X)

    # Fit the regression model
    rdd_model = sm.OLS(y, X).fit()
    print("\nRegression Discontinuity Model Summary:")
    print(rdd_model.summary())

    # Bootstrap ATE Confidence Interval
    ate_bootstrap = []
    for _ in range(n_bootstraps):
        X_sample, y_sample = resample(X, y)
        model = sm.OLS(y_sample, X_sample).fit()
        ate_bootstrap.append(model.params[1])  # Assuming 'treatment' is the second parameter

    ate_ci = np.percentile(ate_bootstrap, [2.5, 97.5])
    print(f"\nATE: {rdd_model.params[1]:.4f}, 95% CI: [{ate_ci[0]:.4f}, {ate_ci[1]:.4f}]")

    # Step 6: Plotting Predicted Values for Each Treatment Group
    plt.figure(figsize=(10, 6))
    plt.scatter(data[distance_col], data[price_col], alpha=0.5, label="Data points")

    for treatment_code, color, label in zip([0, 1, 2], ["blue", "orange", "red"], ["0-0.5 km", "0.5-1 km", "above 1 km"]):
        pred = rdd_model.predict(X[data['treatment'] == treatment_code])
        plt.plot(data[data['treatment'] == treatment_code][distance_col], pred, color=color, label=label)

    # Adding vertical lines to indicate group boundaries
    plt.axvline(0.5, color="green", linestyle="--", label="0.5 km mark")
    plt.axvline(1, color="purple", linestyle="--", label="1 km mark")
    plt.xlabel("Distance to Light Rail (km)")
    plt.ylabel("Apartment Price (NIS)")
    plt.title("Regression Discontinuity Analysis on Apartment Prices")
    plt.legend()
    plt.show()

    # Display regression coefficients table
    results_table = pd.DataFrame({
        'Coefficient': rdd_model.params,
        'Standard Error': rdd_model.bse,
        't-Statistic': rdd_model.tvalues,
        'p-Value': rdd_model.pvalues
    })
    print("\nRegression Coefficients:")
    print(results_table)


# Usage example
distance_column = 'direct_distance_km'
price_column = 'adjusted_sale_value_nis'
numeric_features = ['year_of_construction', 'area', 'bedrooms']

regression_discontinuity_analysis(df_pre_metro, distance_column, price_column, numeric_features)
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.utils import resample

# Load and prepare the data
df = pd.read_csv('df_updated.csv')

# Define the price index adjustment factor
price_index = {
    2021: 1.00,
    2022: 1.147,
    2023: 1.147 * 0.99,
    2024: 1.147 * 0.99 * 1.053
}

df = df[(df['area'] > 2) & (df['percentage_of_sale'] > 0.2)]
df['adjusted_sale_value_nis'] = df.apply(lambda row: row['sale_value_nis'] / (row['percentage_of_sale'] * price_index[row['year']] * row['area']), axis=1)
df_pre_metro = df[(df['year'].isin([2021, 2022, 2023])) & (df['sale_date'] < '2023-08-17')].copy()

# Define distance groups
df_pre_metro.loc[:, 'distance_group'] = pd.cut(
    df_pre_metro['direct_distance_km'],
    bins=[0, 0.5, 1, float('inf')],
    labels=['0-0.5 km', '0.5-1 km', 'above 1 km']
)

df_pre_metro.loc[:, 'treatment'] = df_pre_metro['distance_group'].map({
    '0-0.5 km': 0,
    '0.5-1 km': 1,
    'above 1 km': 2
})

# Covariate adjustment function for group comparison
def compare_groups(data, distance_groups, numeric_features, categorical_features):
    results = {}
    for feature in numeric_features:
        feature_data = []
        for group_name, group_code in distance_groups.items():
            group_values = data[data['distance_group'] == group_code][feature]
            feature_data.append({
                'group': group_name,
                'mean': group_values.mean(),
                'median': group_values.median(),
                'std': group_values.std()
            })
        results[feature] = pd.DataFrame(feature_data)

    for feature in categorical_features:
        feature_data = []
        for group_name, group_code in distance_groups.items():
            value_counts = data[data['distance_group'] == group_code][feature].value_counts(normalize=True)
            feature_data.append({
                'group': group_name,
                'most_common': value_counts.index[0],
                'most_common_pct': value_counts.iloc[0] * 100
            })
        results[feature] = pd.DataFrame(feature_data)

    return results

# Define distance groups and features for comparison
distance_groups = {'0-0.5 km': '0-0.5 km', '0.5-1 km': '0.5-1 km', 'above 1 km': 'above 1 km'}
numeric_features = ['year_of_construction', 'area', 'adjusted_sale_value_nis', 'sale_value_nis', 'bedrooms']
categorical_features = ['nature', 'city', 'closest_direct_address']

# Perform group comparison
comparison_results = compare_groups(df_pre_metro, distance_groups, numeric_features, categorical_features)

# Print comparison results
print("Pre-Metro Group Comparability Analysis (2021-2023):")
print("=" * 50)
for feature, result_df in comparison_results.items():
    print(f"\nFeature: {feature}")
    print(result_df)
    print("-" * 50)

# Statistical tests for numeric features
print("\nStatistical Tests for Numeric Features:")
print("=" * 50)
for feature in numeric_features:
    print(f"\nFeature: {feature}")
    groups = [df_pre_metro[df_pre_metro['distance_group'] == code][feature] for code in distance_groups.values()]
    f_statistic, p_value = stats.f_oneway(*groups)
    print(f"One-way ANOVA: F-statistic = {f_statistic:.4f}, p-value = {p_value:.4f}")

# Chi-square tests for categorical features
print("\nChi-square Tests for Categorical Features:")
print("=" * 50)
for feature in categorical_features:
    print(f"\nFeature: {feature}")
    contingency_table = pd.crosstab(df_pre_metro['distance_group'], df_pre_metro[feature])
    chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
    print(f"Chi-square test: chi2 = {chi2:.4f}, p-value = {p_value:.4f}")

# RDD Analysis with Covariate Adjustment and Bootstrap ATE
def regression_discontinuity_analysis(data, distance_col, price_col, numeric_features, n_bootstraps=1000):
    # Prepare data for regression with dummies for treatment groups
    X = pd.get_dummies(data[['treatment'] + numeric_features], drop_first=True)
    y = data[price_col]

    # Ensure all columns in X and y are numeric, drop NaNs
    X = X.apply(pd.to_numeric, errors='coerce').dropna()
    y = pd.to_numeric(y, errors='coerce').loc[X.index]

    # Add a constant for the intercept
    X = sm.add_constant(X)

    # Fit the main regression model
    rdd_model = sm.OLS(y, X).fit()
    print("\nRegression Discontinuity Model Summary:")
    print(rdd_model.summary())

    # Bootstrap ATE Confidence Interval
    ate_bootstrap = []
    for _ in range(n_bootstraps):
        X_sample, y_sample = resample(X, y)
        model = sm.OLS(y_sample, X_sample).fit()
        ate_bootstrap.append(model.params[1])  # Assuming 'treatment' is the second parameter

    ate_ci = np.percentile(ate_bootstrap, [2.5, 97.5])
    print(f"\nATE: {rdd_model.params[1]:.4f}, 95% CI: [{ate_ci[0]:.4f}, {ate_ci[1]:.4f}]")

    # Group descriptive statistics
    group_stats = data.groupby('distance_group')[price_col].agg(['mean', 'median', 'std']).reset_index()
    print("\nGroup Statistics:")
    print(group_stats)

    # Visualize regression and group differences
    plt.figure(figsize=(10, 6))
    plt.scatter(data[distance_col], data[price_col], color="blue", alpha=0.5, label="Data points")

    # Plot regression predictions for each group
    for treatment_code, color, label in zip([0, 1, 2], ["green", "orange", "red"], ["0-0.5 km", "0.5-1 km", "above 1 km"]):
        group_data = data[data['treatment'] == treatment_code]
        pred = rdd_model.predict(X[data['treatment'] == treatment_code])
        plt.plot(group_data[distance_col], pred, color=color, label=f"Predicted ({label})")

    # Add vertical lines for group boundaries
    plt.axvline(0.5, color="green", linestyle="--", label="0.5 km mark")
    plt.axvline(1, color="purple", linestyle="--", label="1 km mark")
    plt.xlabel("Distance to Light Rail (km)")
    plt.ylabel("Adjusted Apartment Price (NIS)")
    plt.title("Regression Discontinuity Analysis with Predicted Apartment Prices")
    plt.legend()
    plt.show()

    # Display regression coefficients table
    results_table = pd.DataFrame({
        'Coefficient': rdd_model.params,
        'Standard Error': rdd_model.bse,
        't-Statistic': rdd_model.tvalues,
        'p-Value': rdd_model.pvalues
    })
    print("\nRegression Coefficients:")
    print(results_table)

# Define numeric features for covariate adjustment
numeric_features = ['year_of_construction', 'area', 'bedrooms']

# Run the enhanced RDD analysis
distance_column = 'direct_distance_km'
price_column = 'adjusted_sale_value_nis'
regression_discontinuity_analysis(df_pre_metro, distance_column, price_column, numeric_features)
```

# Covariate Adjustment

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.utils import resample

# Load and prepare the data
df = pd.read_csv('df_updated.csv')

# Define the price index adjustment factor
price_index = {
    2021: 1.00,
    2022: 1.147,
    2023: 1.147 * 0.99,
    2024: 1.147 * 0.99 * 1.053
}

df = df[(df['area'] > 2) & (df['percentage_of_sale'] > 0.2)]
df['adjusted_sale_value_nis'] = df.apply(lambda row: row['sale_value_nis'] / (row['percentage_of_sale'] * price_index[row['year']] * row['area']), axis=1)
df_pre_metro = df[(df['year'].isin([2021, 2022, 2023])) & (df['sale_date'] < '2023-08-17')].copy()

# Define distance groups
df_pre_metro.loc[:, 'distance_group'] = pd.cut(
    df_pre_metro['direct_distance_km'],
    bins=[0, 0.5, 1, float('inf')],
    labels=['0-0.5 km', '0.5-1 km', 'above 1 km']
)

df_pre_metro.loc[:, 'treatment'] = df_pre_metro['distance_group'].map({
    '0-0.5 km': 0,
    '0.5-1 km': 1,
    'above 1 km': 2
})


# Enhanced Covariate Adjustment Analysis with variable-wise assessment
def covariate_adjustment_analysis(data, numeric_features, categorical_features, price_col, n_bootstraps=1000):
    # Summary statistics for each variable by treatment group
    print("\nCovariate Adjustment Group Comparability Analysis:")
    for feature in numeric_features + [price_col]:
        group_stats = data.groupby('distance_group')[feature].agg(['mean', 'median', 'std']).reset_index()
        print(f"\nFeature: {feature}")
        print(group_stats)
        print("-" * 50)

    # Perform ANOVA for numeric features
    print("\nStatistical Tests for Numeric Features:")
    for feature in numeric_features:
        groups = [data[data['distance_group'] == group][feature] for group in data['distance_group'].unique()]
        f_stat, p_value = stats.f_oneway(*groups)
        print(f"\nFeature: {feature}")
        print(f"One-way ANOVA: F-statistic = {f_stat:.4f}, p-value = {p_value:.4f}")

    # Chi-square test for categorical features
    print("\nChi-square Tests for Categorical Features:")
    for feature in categorical_features:
        contingency_table = pd.crosstab(data['distance_group'], data[feature])
        chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)
        print(f"\nFeature: {feature}")
        print(f"Chi-square test: chi2 = {chi2:.4f}, p-value = {p_value:.4f}")

    # Prepare the design matrix X with treatment and numeric features
    X = pd.get_dummies(data[['treatment'] + numeric_features], drop_first=True)
    y = data[price_col]

    # Ensure all columns in X and y are numeric and drop NaN values
    X = X.apply(pd.to_numeric, errors='coerce').dropna()
    y = pd.to_numeric(y, errors='coerce').loc[X.index]

    # Add a constant for the intercept
    X = sm.add_constant(X)

    # Fit the main regression model
    cov_adj_model = sm.OLS(y, X).fit()
    print("\nCovariate Adjustment Model Summary:")
    print(cov_adj_model.summary())

    # Bootstrap ATE Confidence Interval
    ate_bootstrap = []
    for _ in range(n_bootstraps):
        X_sample, y_sample = resample(X, y)
        model = sm.OLS(y_sample, X_sample).fit()
        ate_bootstrap.append(model.params[1])  # Assuming 'treatment' is the second parameter

    ate_ci = np.percentile(ate_bootstrap, [2.5, 97.5])
    print(f"\nATE: {cov_adj_model.params[1]:.4f}, 95% CI: [{ate_ci[0]:.4f}, {ate_ci[1]:.4f}]")

    # Visualize Predictions and Observed Data
    preds = cov_adj_model.predict(X)

    plt.figure(figsize=(10, 6))
    plt.scatter(data['direct_distance_km'], data[price_col], color="blue", alpha=0.7, marker="o", label="Data points", edgecolor="k")
    plt.scatter(data['direct_distance_km'], preds, color="red", alpha=0.4, marker="x", label="Adjusted predictions")
    plt.axvline(0.5, color="green", linestyle="--", label="0.5 km mark")
    plt.axvline(1, color="purple", linestyle="--", label="1 km mark")
    plt.xlabel("Distance to Light Rail (km)")
    plt.ylabel("Adjusted Apartment Price (NIS)")
    plt.title("Covariate Adjustment Analysis with Predicted Apartment Prices")
    plt.legend()
    plt.show()

    # Generate and display regression coefficients table
    results_table = pd.DataFrame({
        'Coefficient': cov_adj_model.params,
        'Standard Error': cov_adj_model.bse,
        't-Statistic': cov_adj_model.tvalues,
        'p-Value': cov_adj_model.pvalues
    })
    print("\nCovariate Adjustment Regression Coefficients:")
    print(results_table)

# Define numeric and categorical features
numeric_features = ['year_of_construction', 'area', 'bedrooms', 'adjusted_sale_value_nis']
categorical_features = ['nature', 'city', 'closest_direct_address']
price_column = 'adjusted_sale_value_nis'

# Run the Covariate Adjustment Analysis with variable-wise assessment
covariate_adjustment_analysis(df_pre_metro, numeric_features, categorical_features, price_column)
```
